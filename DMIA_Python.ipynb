# Numpy

1) numpy.linspace

```python
def linsp_list_comprehension(begin, end, num):
    step = (float(end) - begin)/(num-1)
    ar = [begin+step*i for i in range(num)]
    return step, ar
```

```python
def linsp_cycle(begin, end, num):
    step = (float(end) - begin)/(num-1)
    ar = []
    for i in range(num):
        ar.append(begin+step*i)
    return step, ar
```

```python
import numpy as np
```

```python
%time linsp_cycle(1,3,1000)[1]
```

```python
%time linsp_list_comprehension(1,3,1000)[1]
```

```python
%time np.linspace(1,3,1000)
```

```python
%time sum(linsp_cycle(1,3,1000)[1])
```

# Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
f, a = plt.subplots(1,2)
x = np.linspace(0.1,8,100)
y_1 = x
y_2 = np.sin(x)
y_3 = np.sqrt(x)
a[0].plot(x,y_1,label='$x$')
a[0].plot(x,y_2,label='$\sin(x)$')
a[0].plot(x,y_3,label='$\sqrt{x}$')
a[0].set_ylabel('$y$')
a[0].set_xlabel('$x$')
a[0].legend(loc='upper left')
y_4 = np.cos(x)
y_5 = np.log(x)
a[1].plot(x,y_4,label='$\cos(x)$')
a[1].plot(x,y_5,label='$\ln(x)$')
a[1].set_ylabel('$y$')
a[1].set_xlabel('$x$')
a[1].legend(loc='upper right')
plt.savefig('python')
```
